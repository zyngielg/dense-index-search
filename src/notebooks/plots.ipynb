```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
input_data = {
    "IR-ES retriever + BERT reader": [
        (1.3917847471924785, 1.3809429109096527),
        (1.384761938107051, 1.3715861141681671),
        (1.3681087512581327, 1.3540987998247147),
        (1.325900379393168, 1.366346225142479)  
    ] 
}
```

```python
def create_training_plot(train_loss, val_loss, title):
    assert len(train_loss) == len(val_loss)
    epochs = list(range(len(train_loss)))
    fig, ax = plt.subplots(figsize=(4, 4))
    train, = ax.plot(epochs, train_loss, color="blue", marker="o")
    val, = ax.plot(epochs, val_loss, color="orange", marker="o")
    
    ax.legend([train, val],["Train loss", "Validation loss"])
    ax.set_xlabel("Epochs")
    ax.set_ylabel("Loss")
    plt.xticks(epochs)
    
    plt.show()
```

```python
for title, values in input_data.items():
    print(title)
    train_loss = [x[0] for x in values]
    val_loss = [x[1] for x in values]
    create_training_plot(train_loss, val_loss, title)
```
